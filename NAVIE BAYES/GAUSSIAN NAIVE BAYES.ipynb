```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
from sklearn.datasets import make_classification
```

```python
X,Y=make_classification(n_samples=200,n_features=2,n_redundant=0,n_informative=2,random_state=5)
```

```python
plt.style.use("seaborn")
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
gnb.fit(X,Y)
```

```python
gnb.score(X,Y)
```

```python
a=gnb.predict(X)
print(a)
```

```python
print(Y)
```

```python
acc=np.sum(a==Y)/X.shape[0]
print(acc)
```
